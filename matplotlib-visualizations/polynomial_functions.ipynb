```python
#Na jednym rysunku wykonaj wykresy funkcji x, x2 + 1, x3 + 2 na przedziale
#[−2, 2] wraz z legendą.

import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-2,2)
y=x
plt.plot(x,y, label='f(x)=x')
plt.plot(x,x**2+1,label='f(x)=x**2+1')
plt.plot(x,x**3+1,label='f(x)=x**3+1')
plt.legend()
```
