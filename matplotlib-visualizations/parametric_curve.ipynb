```python
#Napisz funkcję, która dla podanego a > 0 wykonuje wykres krzywej parametrycznej φ(t) = [a cos3(t), a sin3(t)].
#Wykorzystując tą funkcję narysuj wykres, gdzie znajdą się takie krzywe dla a = 1, 2, 3, 4, 5.

import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0, 2*np.pi, 500)
for a in range(1,6):
    plt.plot(a*np.sin(t)**3,a*np.cos(t)**3)

```
